```python
import numpy as np
import pandas as pd
```

```python
rainfall = pd.read_csv('Seattle2014.csv')['PRCP'].values
inches = rainfall / 254 # 1/10mm -> inches
# inches = rainfall / 254: This converts the rainfall data from tenths of millimeters (the common unit in many datasets) to inches. 
# Since there are 254 tenths of a millimeter in an inch, dividing by 254 gives you the value in inches.
inches.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn;
```

```python
plt.hist(inches, 40);
```

```python
print("Number days without rain: ", np.sum(inches == 0))
print("Number days with rain: ", np.sum(inches != 0))
print("Days with more than 0.5 inches:", np.sum(inches > 0.5))
print("Rainy days with < 0.1 inches :", np.sum((inches > 0) & (inches < 0.2)))
```
