<h1 style="color:blue">Reshape dataframe using stack/unstack</h1>

```python
import pandas as pd
df = pd.read_excel("stocks.xlsx",header=[0,1])
df
```

```python
df.stack()
# converts columns to rows and rows to columns
```

```python
df.stack(level=0)
# Converts same column to row  
```

```python
df_stacked=df.stack()
df_stacked
```

```python
df_stacked.unstack()
```
