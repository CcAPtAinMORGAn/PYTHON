```python
import math
from scipy.stats import norm

# Given data
n = 50
PopulationMean = 100
SampleMean = 110
StandardDeviation = 15

# Step 1: Calculate the Z-test value
Z_test = (SampleMean - PopulationMean) / (StandardDeviation / math.sqrt(n))
print("Z-Test value is", Z_test)

# Step 2: Define significance level and find critical Z-value from the Z-table
alpha = 0.05

# For a one-tailed test at alpha = 0.05, we look for the 95th percentile (1 - alpha)
critical_value = norm.ppf(1 - alpha)
print("Critical Z-value from the Z-table is", critical_value)

# Step 3: Compare Z-test value with critical value
if Z_test > critical_value:
    print("Null Hypothesis is rejected")
else:
    print("Null Hypothesis is accepted")

```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
n = 50
PopulationMean = 100
SampleMean = 110
StandardDeviation = 15

# Step 1: Calculate the Z-test value
Z_test = (SampleMean - PopulationMean) / (StandardDeviation / math.sqrt(n))
print("Z-Test value is", Z_test)

# Step 2: Define significance level and find critical Z-value from the Z-table
alpha = 0.05

# For a one-tailed test at alpha = 0.05, we look for the 95th percentile (1 - alpha)
critical_value = norm.ppf(1 - alpha)
print("Critical Z-value from the Z-table is", critical_value)

# Step 3: Compare Z-test value with critical value
if Z_test > critical_value:
    print("Null Hypothesis is rejected")
else:
    print("Null Hypothesis is accepted")

# Plotting the normal distribution curve
x = np.linspace(PopulationMean - 4 * (StandardDeviation / math.sqrt(n)),
                PopulationMean + 4 * (StandardDeviation / math.sqrt(n)), 1000)
y = norm.pdf(x, PopulationMean, StandardDeviation / math.sqrt(n))

plt.figure(figsize=(10, 6))

# Plot the normal distribution curve
plt.plot(x, y, label='Normal Distribution', color='blue')

# Shade the critical region
plt.fill_between(x, 0, y, where=(x > critical_value), color='red', alpha=0.5, label='Critical Region')

# Mark the Z-test value
plt.axvline(Z_test, color='green', linestyle='--', label=f'Z-Test = {Z_test:.2f}')

# Mark the critical Z-value
plt.axvline(critical_value, color='purple', linestyle='--', label=f'Critical Z = {critical_value:.2f}')

# Labels and title
plt.title('Z-Test Visualization')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend(loc='upper left')

# Show the plot
plt.grid(True)
plt.show()
```
