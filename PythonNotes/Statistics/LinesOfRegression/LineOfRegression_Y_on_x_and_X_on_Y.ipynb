```python
import matplotlib.pyplot as plt
import math

# Given data
x = [78, 36, 98, 25, 75, 82, 90, 62, 65, 39]
y = [84, 51, 91, 60, 68, 62, 86, 58, 53, 47]

# The rest of your code remains unchanged...

x_given_value = 50
y_given_value = 90
n = len(x)

# ------------------- Mean of array x
a = 0
for i in range(n):
    a += x[i] / n

# ------------------ Mean of array y
b = 0
for l in range(n):
    b += y[l] / n

Mean_x = round(a)
print("Mean of x array is ", Mean_x)

Mean_y = round(b)
print("Mean of y array is ", Mean_y, "\n\n")

# -------------------------------------- x_Minus_Mean and y_Minus_Mean
x_Minus_Mean = []
for u in range(n):
    x_Minus_Mean_Logic = x[u] - Mean_x
    x_Minus_Mean.append(x_Minus_Mean_Logic)
print("x minus mean of x array is", x_Minus_Mean)

y_Minus_Mean = []
for z in range(n):
    y_Minus_Mean_Logic = y[z] - Mean_y
    y_Minus_Mean.append(y_Minus_Mean_Logic)
print("y minus mean of y array is", y_Minus_Mean, "\n\n")

# --------------------------- summation x square
x_square_array = []

for j in range(n):
    x_square = (x[j] - Mean_x)**2
    x_square_array.append(x_square)

print("(x minus mean of x) square array is ", x_square_array)

Summision_x_square = sum(x_square_array)
print("Summision of (x - Mean of x)^2 is", Summision_x_square, "\n\n")

# --------------------------- summation y square
y_square_array = []
for k in range(n):
    y_square = (y[k] - Mean_y)**2
    y_square_array.append(y_square)
print("(y minus mean of y) square array is", y_square_array)

Summision_y_square = sum(y_square_array)
print("Summision of (y - Mean of y)^2 is", Summision_y_square, "\n\n")

# -------------------------- summation x * y
summision_appended_xy_array = []
for m in range(n):
    appended_xy = x_Minus_Mean[m] * y_Minus_Mean[m]
    summision_appended_xy_array.append(appended_xy)
print("Array of x * y is", summision_appended_xy_array)

summision_appended_xy = sum(summision_appended_xy_array)
print("Summision of (x - mean of x) * (y - mean of y) is", summision_appended_xy, "\n\n")

# ---------------- line of regression y on x
byx = summision_appended_xy / Summision_x_square
print("The value of byx is", byx)

y_unknown = (byx * (x_given_value - Mean_x)) + Mean_y

# --------------- line of regression x on y
bxy = summision_appended_xy / Summision_y_square
print("The value of bxy is", bxy)

x_unknown = (bxy * (y_given_value - Mean_y)) + Mean_x

print("******************************************************************")

print("The value of y is", y_unknown)
print("The value of x is", x_unknown)

print("******************************************************************")
# ----------------------- Plotting section
plt.figure(figsize=(10, 6))

# Plot the original data points
plt.scatter(x, y, color='blue', label='Data points')

# Plot the regression line y on x (y = byx * (x - Mean_x) + Mean_y)
x_line = [min(x), max(x)]
y_line = [byx * (xi - Mean_x) + Mean_y for xi in x_line]
plt.plot(x_line, y_line, color='red', label='Regression line y on x')

# Plot the regression line x on y (x = bxy * (y - Mean_y) + Mean_x)
y_line2 = [min(y), max(y)]
x_line2 = [bxy * (yi - Mean_y) + Mean_x for yi in y_line2]
plt.plot(x_line2, y_line2, color='green', label='Regression line x on y')

# Plot the predicted y for given x value (x_given_value)
plt.scatter(x_given_value, y_unknown, color='orange', zorder=5, label=f'Predicted y for x={x_given_value}')

# Plot the predicted x for given y value (y_given_value)
plt.scatter(x_unknown, y_given_value, color='purple', zorder=5, label=f'Predicted x for y={y_given_value}')

# Add labels and title
plt.title('Linear Regression Analysis')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
