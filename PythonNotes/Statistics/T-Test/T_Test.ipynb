```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as stats

# Given values
n = 25
PopulationMean = 45
SampleMean = 75
StandardDeviation = 25
alpha = 0.05
df = n - 1  # Degree of Freedom

# Calculate the T-statistic
T_test = (SampleMean - PopulationMean) / (StandardDeviation / math.sqrt(n))

# Critical t-values
critical_value_two_tailed = stats.t.ppf(1 - alpha / 2, df)
critical_value_one_tailed = stats.t.ppf(1 - alpha, df)

# Print the calculated t-statistic and critical values
print("T-test value is ", T_test)
print("Critical t-value for two-tailed test: ", critical_value_two_tailed)
print("Critical t-value for one-tailed test: ", critical_value_one_tailed)

# Plot the T-distribution curve
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(10, 6))

# Plot the T-distribution
plt.plot(x, y, label=f'T-distribution (df={df})', color='blue')

# Shade the critical region for two-tailed test
plt.fill_between(x, y, where=(x >= critical_value_two_tailed) | (x <= -critical_value_two_tailed), 
                 color='red', alpha=0.3, label='Critical Region (Two-Tailed)')

# Plot the T-test statistic
plt.axvline(T_test, color='green', linestyle='dashed', linewidth=2, label=f'T-test Value: {T_test:.2f}')

# Add labels and title
plt.title('T-distribution with Critical Region and T-test Value')
plt.xlabel('T-statistic')
plt.ylabel('Probability Density')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()

# Hypothesis Test Decision
if abs(T_test) > critical_value_two_tailed:
    print("Since", T_test, "is greater than", critical_value_two_tailed, "Null hypothesis is rejected")
else:
    print("Since", T_test, "is lesser than", critical_value_two_tailed, "Null hypothesis is accepted")
```
