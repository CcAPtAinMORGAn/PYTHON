```python
# Variance
# Variance is another number that indicates how spread out the values are.

# In fact, if you take the square root of the variance, you get the standard deviation!

# Or the other way around, if you multiply the standard deviation by itself, you get the variance!

# To calculate the variance you have to do as follows:

# 1. Find the mean:

# (32+111+138+28+59+77+97) / 7 = 77.4

# 2. For each value: find the difference from the mean:

#  32 - 77.4 = -45.4
# 111 - 77.4 =  33.6
# 138 - 77.4 =  60.6
#  28 - 77.4 = -49.4
#  59 - 77.4 = -18.4
#  77 - 77.4 = - 0.4
#  97 - 77.4 =  19.6

# 3. For each difference: find the square value:

# (-45.4)2 = 2061.16
#  (33.6)2 = 1128.96
#  (60.6)2 = 3672.36
# (-49.4)2 = 2440.36
# (-18.4)2 =  338.56
# (- 0.4)2 =    0.16
#  (19.6)2 =  384.16

# 4. The variance is the average number of these squared differences:

# (2061.16+1128.96+3672.36+2440.36+338.56+0.16+384.16) / 7 = 1432.2

```

```python
# Use the NumPy var() method to find the variance:

import numpy as np

speed = [32,111,138,28,59,77,97]

x = np.var(speed)

print("Variance is",x)
```

```python
import math


# Data
speed = [32, 111, 138, 28, 59, 77, 97]

# Calculate mean
n = len(speed)
Mean = sum(speed) / n
print("Mean is", Mean)

# Calculate standard deviation
z = sum((x - Mean) ** 2 for x in speed)
Standard_Deviation = math.sqrt(z / n)
print("Standard Deviation is", Standard_Deviation)

Variance=(Standard_Deviation)**2
print("Variance is", Variance)
```

```python
import math
import matplotlib.pyplot as plt

# Data
speed = [32, 111, 138, 28, 59, 77, 97]

# Calculate mean
n = len(speed)
Mean = sum(speed) / n
print("Mean is", Mean)

# Calculate standard deviation
z = sum((x - Mean) ** 2 for x in speed)
Standard_Deviation = math.sqrt(z / n)
print("Standard Deviation is", Standard_Deviation)

# Calculate variance
Variance = Standard_Deviation ** 2
print("Variance is", Variance)

# Plotting
plt.figure(figsize=(10, 6))

# Plot the original data points as a scatter plot
plt.scatter(range(1, n + 1), speed, color='b', label='Data Points', zorder=5)

# Plot the mean as a horizontal red dashed line
plt.axhline(y=Mean, color='r', linestyle='--', label=f'Mean: {Mean:.2f}')

# Plot the standard deviation lines (mean  standard deviation)
plt.axhline(y=Mean + Standard_Deviation, color='g', linestyle='--', label=f'Mean + SD: {Mean + Standard_Deviation:.2f}')
plt.axhline(y=Mean - Standard_Deviation, color='g', linestyle='--', label=f'Mean - SD: {Mean - Standard_Deviation:.2f}')

# Highlight variance area (between mean - SD and mean + SD)
plt.fill_between([0, n + 1], Mean - Standard_Deviation, Mean + Standard_Deviation, color='lightgreen', alpha=0.5, label=f'Variance Area')

# Title and labels
plt.title('Data with Mean, Standard Deviation, and Variance')
plt.xlabel('Index')
plt.ylabel('Speed')

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
